In this script, we explored the 2 open-source datasets predicted by the Streetscore project and created a target variable for our project to use.

From this script, we can see that the Streetscore project generated a 'q-score' label (based on the Place Pulse project which originally has 3 labels, as we explored in PlacePulse1.0_explore.ipynb).

We generated a binary 'safety' label based on the 'q-score' label. We define safety = 0 stands for 'less safe' and safety = 1 stands for 'more safe'.


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
boston = pd.read_csv("~/Desktop/ML1030/streetscore_dataset/streetscore_boston.csv")
boston['city'] = 'Boston'
ny = pd.read_csv("~/Desktop/ML1030/streetscore_dataset/streetscore_newyorkcity.csv")
ny['city'] = 'New York'
```

```python
print(boston.shape)
print(ny.shape)
```

```python
df = pd.concat([boston, ny])
```

```python
sns.distplot(df['q-score'], bins=10, kde=True, rug=False)
#plt.plot([df['q-score'].median(), df['q-score'].median()], [0, 0.1])
```

```python
sns.distplot(df['q-score'], bins=2, kde=True, rug=False)
```

```python
df['binof2'] = pd.qcut(df['q-score'], 2)
```

```python
df['binof2_class'] = pd.qcut(df['q-score'], 2, labels = range(2))
```

```python
df['binof2'].value_counts().sort_index().plot(kind='bar')
```

```python
df['binof2'].value_counts().sort_index()
```

```python
df['binof2_class'].value_counts().sort_index()
```

```python
df['safety'] = np.where(df['binof2_class'] == 0 , 0, 1)
```
