In this script, we 1). check how many images were fetched; 2) merge the fetched images with target variable.

1). After we used fetchimage.py to fetch images using Google Street View Static API, we found that for the 20,000 samples, there were less than 100 images not available using the Google Street View Static API.

We ended up getting around 19900 images altogether. Their portion of 'safety' and 'subdistrict' were not affected.

In this script, we record the results while fetching 20,000 images altogether. Note that while fetching train and test images separately, the examining codes would be similar and the portion of the 'safety' and 'subdistrict' were not affected as well.

2). We then merged the fetched images' metadata with the target variable. The produced table lists images that we were able to fetch, and each sample had a target variable.

In another script, merge_fetched_train_test_image_with_target.ipynb, we conduct this operation with train / test images separately. You may consider to use one of these 2 scripts, depending on you are saving the images to one folder or two (train / test separately) folders.

#### Check whether the fetched images' portion of 'safety' and 'subdistrict' were affected

Method: Merge table 'image_fetched' with table 'boston_safety_subsample.csv' by index. 

Result: the 19904 samples' portion of 'safety' and 'subdistrict' were not affected comparing with that of the original 'boston_safety_subsample.csv'

```python
boston_safety_subsample = pd.read_csv("~/Desktop/ML1030/us_safety/boston_safety_subsample.csv")
```

```python
boston_safety_subsample.head()
```

```python
boston_safety_subsample.shape
```

```python
image_fetched = metadata[metadata['status'] == 'OK']
```

```python
image_fetched.shape
```

```python
image_fetched_with_target = pd.merge(image_fetched, boston_safety_subsample, left_index=True, right_index=True)
```

```python
image_fetched_with_target.head()
```
